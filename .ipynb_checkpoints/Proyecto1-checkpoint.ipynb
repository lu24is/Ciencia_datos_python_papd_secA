# Zeramiks S.A

```python
import json
import random

def generar_codigo():
    codigo = ''
    suma = 0
    categoria = ''

    for i in range(0, 7):
        letra = chr(random.randint(65, 90))  # Convertimos número a carácter ASCII
        valor = random.randint(0, 9)
        suma += valor
        codigo += letra + str(valor)

    if suma >= 35:
        categoria = 'A'
    elif 16 <= suma < 35:
        categoria = 'B'
    else:
        categoria = 'C'

    return codigo, categoria

def guardar_codigo(codigo, categoria):
    with open("codigos.json", "a") as archivo:
        json.dump({"codigo": codigo, "categoria": categoria}, archivo)
        archivo.write('\n')

def crear_codigos():
    cantidad = int(input("Ingrese la cantidad de códigos que desea generar: "))
    for _ in range(cantidad):
        codigo, categoria = generar_codigo()
        guardar_codigo(codigo, categoria)
        print(f"Código generado: {codigo}, Categoría: {categoria}")

def ver_codigos():
    categoria = input("Ingrese la categoría de los códigos que desea ver (A/B/C): ")

    try:
        with open("codigos.json", "r") as archivo:
            for linea in archivo:
                data = json.loads(linea)
                if data['categoria'] == categoria:
                    print(data)
                else:
                    print("No hay codigos almacenados con esa categoría")
    except FileNotFoundError:
        print("No hay códigos almacenados.")

def menu():
    nombre_usuario = input("Ingrese su nombre: ")
    print(f"Bienvenido, {nombre_usuario}!")

    while True:
        print("\nMenú:")
        print("1. Crear códigos nuevos.")
        print("2. Ver códigos creados.")
        print("3. Salir.")

        opcion = input("Seleccione una opción: ")

        if opcion == "1":
            crear_codigos()
        elif opcion == "2":
            ver_codigos()
        elif opcion == "3":
            print("¡Hasta luego! El programa se cerrará")
            break
        else:
            print("Opción inválida. Por favor, seleccione una opción válida.")

menu()
```
